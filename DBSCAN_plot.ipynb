```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import DBSCAN


%matplotlib inline
```

```
# Import dataset using pandas

data = pd.read_csv("Complex8_N15.csv")
data.tail()
```

```
MIN_SAMPLES = 5
EPSILON = 5

clustering = DBSCAN(eps=EPSILON, min_samples=MIN_SAMPLES).fit(data)
clustering_label = clustering.fit_predict(data)

fig,ax = plt.subplots(figsize=(7, 7), dpi=100)
ax.scatter(data['V1'], data['V2'], c=clustering_label, alpha=0.25, s=60, linewidths=0)
ax.set_title('Cluster Found DBSCAN(eps=5)')
```
